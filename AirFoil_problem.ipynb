```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
import random

columns = ['Frequency','Angle_of_attack', 'Chord_length', 'Free_stream_velocity','Suction_side_displacement_thickness','Scaled_sound_pressure_level']
features = ['Frequency','Angle_of_attack', 'Chord_length', 'Free_stream_velocity','Suction_side_displacement_thickness']

data = pd.read_csv('airfoil_self_noise.dat', sep='\t', header=None, names=columns)
data.head()
```

```python
correlation = data.corr()
correlation.head()
```

```python
#splitting between testing and training data sets
# 20% for testing

train=data.sample(frac=0.8) #random state is a seed value
test=data.drop(train.index)

train.head()
```

```python
train_labels = train["Scaled_sound_pressure_level"].copy()
train_data = train.drop('Scaled_sound_pressure_level',axis=1)

test_labels = test["Scaled_sound_pressure_level"].copy()
test_data = test.drop('Scaled_sound_pressure_level',axis=1)
test_data.head()
```

```python
X=np.matrix.tolist(np.asmatrix(train_data))
Y=np.array(train_labels).reshape(-1,1)

X1 = []       #for scaling will be reverted back
#SCALING
for i in np.transpose(X):
    i  = (i-i.min())/(i.max()-i.min())     
    X1.append(i)
    
X=np.transpose(X1)     #x after scaling
X = np.transpose(np.hstack((np.ones(len(X)).reshape(-1,1),X))) #adding 1 
w=np.ones(X.shape[0])
# w = np.random.normal(0,1,X.shape[0])

Y_estimated = np.matmul(np.transpose(w),X).reshape(-1,1)
print(Y_estimated)

#shape of X = (6,1202)
#shape of w = (6,1)
#shape of Y or Y_estimated = (1202,1)
```

```python
#Error
def erro(Y1,Y2):
    m=len(Y1)
    a=0
    for i in range(m):
        a=a+((Y1[i] - Y2[i])**2)[0]
    return a*(1/(2*m))

print(erro(Y_estimated,Y))
```

```python
#Gradient Descent
scaling_factor = 0.0001 * len(Y_estimated)
i=0

while(i<10000 and erro(Y_estimated,Y)>0.2):
    w= w - scaling_factor * np.matmul(X,(Y_estimated - Y))[0]
    i=i+1
    
    Y_estimated=np.matmul(np.transpose(w),X).reshape(-1,1)
    print(i)
    if i<5:
        print (erro(Y_estimated,Y))
print(w)
```
